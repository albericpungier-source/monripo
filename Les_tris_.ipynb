```python
def fusion(L,M):
    S=[]
    while len(M)!=0 and len(L)!=0:
        if L[-1]<M[-1]:
            S.append(M.pop())
        else:
            S.append(L.pop())
    while len(M)!=0:
        S.append(M.pop())
    while len(L)!=0:
        S.append(L.pop())
#    S.reverse()
    return S

def tri_fusion(L):
    n=len(L)
    if n==1 or n==0:
        return L
    m=n//2
    return fusion(tri_fusion(L[:m]),tri_fusion(L[m:]))
        
```
