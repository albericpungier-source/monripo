```python
from collections import deque

def BFS(i,L):
    d=deque()
    Vis=[i]
    for k in L[i-1]:
        d.append(k)
    while len(d)!=0:
        m=d.popleft()
        Vis.append(m)
        for elt in L[m-1]:
            if elt not in Vis and elt not in d:
                d.append(elt)
    return Vis
```

```python
def BFS_correction_incomplète(L, source):
    l= len(L)
    visité=[False for i in range(l)]
    parent=[None for n in range(l)]
    F=deque()
    visité[source]=True
    F.append(source)
    while len(F)!=0:
        u=F.popleft()
        for v in L:
            if visité[v]==False:
                visité[v]=True
                parent[v] = u
                F.append(v)
    return parent
```
