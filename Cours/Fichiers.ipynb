### Lecture, √©criture, ajout

Un fichier permet de stocker des donn√©es durablement, l√† o√π les objets Python n'existent que tant que l'interpr√©teur est ouvert. Pour traiter avec Python des donn√©es stock√©es dans un fichier, ou pour enregistrer dans un fichier le r√©sultat d'un traitement, il nous faut un lien entre les deux. Ce lien est mat√©rialis√© par la fonction `open()`, que l'on fera pr√©c√©d√©rer syst√©matiquement du mot-cl√© `with`.

La fonction `open()` a deux arguments obligatoires : le chemin vers le fichier cibl√© (relatif ou absolu, voir plus haut) et son mode d'ouverture.

Il existe quatre modes, mais trois seulement sont r√©ellement utiles.

- `r` pour `read` : lecture seule des donn√©es stock√©es.
- `w` pour `write` : √©crase le fichier si il existait, puis le recr√©e, vierge, pour √©crire dessus.
- `a` pour `append` : prend un fichier qui existe d√©j√† et √©cris les donn√©es souhait√©es √† la fin.

Une fois le fichier captur√© par le `open()`, on peut lire/√©crire dedans avec les m√©thodes `.read()` et `write()`, selon le mode d'ouverture du fichier.

````python
with open('./asterix', 'w') as file:
    file.write("Je suis un sanglier.")
````

````python
with open('./asterix', 'w') as file:
    file.write("Tu es un sanglier.")
````

````python
with open('./asterix', 'a') as file:
    file.write("\nNous sommes des sangliers.")
````

````python
with open('./asterix', 'r') as file:
    contenu = file.read()
    print(contenu)
````
