```python
class De:
    def __init__(self, up, front, right):
        self.up = up
        self.front = front
        self.right = right
        self.down = 7 - up
        self.back = 7 - front
        self.left = 7 - right

    def update_faces(self):
        self.down = 7 - self.up
        self.back = 7 - self.front
        self.left = 7 - self.right

    def rotate_right(self):
        self.up, self.right = 7 - self.right, self.up
        self.update_faces()

    def rotate_left(self):
        self.up, self.right = self.right, 7 - self.up
        self.update_faces()

    def rotate_down(self):
        self.up, self.front = 7 - self.front, self.up
        self.update_faces()

    def rotate_up(self):
        self.up, self.front = self.front, 7 - self.up
        self.update_faces()

    def egalite(self, d2):
        return (self.right, self.up, self.front) == (d2.right, d2.up, d2.front)


def nombre_toursbis():
    D = De(1, 2, 3)
    for i in range(3):
        D.rotate_right()
    for i in range(3):
        D.rotate_down()
    for i in range(3):
        D.rotate_left()
    for i in range(3):
        D.rotate_up()
    
    t = 1
    while not D.egalite(De(1, 2, 3)):
        t += 1
        for i in range(3):
            D.rotate_right()
        for i in range(3):
            D.rotate_down()
        for i in range(3):
            D.rotate_left()
        for i in range(3):
            D.rotate_up()
    
    return t
```
